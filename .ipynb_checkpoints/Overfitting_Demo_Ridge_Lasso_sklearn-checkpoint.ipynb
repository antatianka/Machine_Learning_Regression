```python
import math
import random
import numpy as np
import pandas as pd
from sklearn import  linear_model
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.linear_model import Ridge
from sklearn.cross_validation import KFold
from sklearn.linear_model import RidgeCV
```

```python
#Create random values for x in interval [0,1)

random.seed(98103)
n = 30
x = pd.Series([random.random() for i in range(n)])
x = x.order()
```

```python
#Compute y
y = x.apply(lambda x: math.sin(4*x))
```

```python
#Add random Gaussian noise to y
random.seed(1)
e = pd.Series([random.gauss(0,1.0/3.0) for i in range(n)])

y = y +e
```

```python
data = pd.DataFrame({'X1':x,'Y':y})
data.sort('X1')
```

```python
#Create function for plotting
def plot_data(data):    
    plt.plot(data['X1'],data['Y'],'k.')
    plt.xlabel('x')
    plt.ylabel('y')

plot_data(data)
```

```python
#Define a function to create  features for a polynomial regression model of any degree:
def polynomial_features(data, deg):
    data = data[['X1']]
    
    data_copy=data.copy()
    
    for i in range(1,deg):
        data_copy['X'+str(i+1)]=data_copy['X'+str(i)]*data_copy['X1']
        #print data_copy['X'+str(i+1)]
    predictors = data_copy.columns.values 
    features = data_copy[predictors]
    return features
```

```python
#Define a function to fit a polynomial linear regression model of degree "deg" to the data in "data":
def polynomial_regression(data, deg):
    target = data['Y']
    
    #print features
    model = linear_model.LinearRegression()
    model.fit(polynomial_features(data, deg), target)
    #print model
   
    return model
```

```python
#Define function to plot data and predictions made, since we are going to use it many times.
def plot_poly_predictions(data, model):
    plot_data(data)

    # Get the degree of the polynomial
    deg = len(model.coef_)
    
    # Create 200 points in the x axis and compute the predicted value for each point
    x_pred = pd.DataFrame({'X1':[i/200.0 for i in range(200)]})
    
    y_pred = model.predict(polynomial_features(x_pred,deg))
    
    # plot predictions
    plt.plot(x_pred['X1'], y_pred, 'g-', label='degree ' + str(deg) + ' fit')
    plt.legend(loc='upper left')
    plt.axis([0,1,-1.5,2])
```

```python
#Create a function that prints the polynomial coefficients in a pretty way :)
def print_coefficients(model):    
    # Get the degree of the polynomial
    #print model.coef_
    deg = len(model.coef_)
    
    # Get learned parameters as a list
    w = [model.intercept_] + list(model.coef_) 

    # Numpy has a nifty function to print out polynomials in a pretty way
    # (We'll use it, but it needs the parameters in the reverse order)
    print 'Learned polynomial for degree ' + str(deg) + ':'
    w.reverse()
    print np.poly1d(w)
 
```

```python
#Fit  degree-2 polynomial to the data generated above:
model = polynomial_regression(data, 2)
```

```python
print_coefficients(model)
```

```python
plot_poly_predictions(data,model)
```

```python
#Fit a degree-4 polynomial
model = polynomial_regression(data, deg=4)
print_coefficients(model)
plot_poly_predictions(data,model)
```

```python
#Fit a degree-16 polynomial
model = polynomial_regression(data, deg=16)
print_coefficients(model)
```

```python
plot_poly_predictions(data,model)
```

## Ridge Regression

```python
#Define our function to solve the ridge objective for a polynomial regression model of any degree:
def polynomial_ridge_regression(data, deg, alpha):
    target = data['Y']
   
    data_copy= polynomial_features(data,deg)
    predictors = data_copy.columns.values
    features =  data_copy[predictors]
    
    
    model = Ridge(alpha)
    
    model.fit(features, target)
  
    return model


```

```python
#Perform a ridge fit of a degree-16 polynomial using a very small penalty strength(alpha)

model_ridge1 = polynomial_ridge_regression(data, deg=16, alpha =1e-25)
print_coefficients(model)
```

```python
plot_poly_predictions(data,model_ridge1)
```

```python
#Perform a ridge fit of a degree-16 polynomial using a very large penalty strength(alpha)
model_ridge2 = polynomial_ridge_regression(data, deg=16, alpha=100)
print_coefficients(model)
```

```python
plot_poly_predictions(data,model_ridge2)
```

```python
#fits for a sequence of increasing lambda values
for alpha in [1e-25, 1e-10, 1e-6, 1e-3, 1e2]:
    model = polynomial_ridge_regression(data, deg=16, alpha=alpha)
    print 'lambda = %.2e' % alpha
    print_coefficients(model)
    print '\n'
    plt.figure()
    plot_poly_predictions(data,model)
    plt.title('Ridge, lambda = %.2e' % alpha)
```

```python
def polynomial_dataframe(feature, degree): # feature is pandas.Series type
    # assume that degree >= 1
    # initialize the dataframe:
    poly_dataframe = pd.DataFrame()
    # and set poly_dataframe['power_1'] equal to the passed feature
    poly_dataframe['power_1'] = feature
    # first check if degree > 1
    if degree > 1:
        # then loop over the remaining degrees:
        for power in range(2, degree+1):
            # first we'll give the column a name:
            name = 'power_' + str(power)
            # assign poly_dataframe[name] to be feature^power; use apply(*)
            poly_dataframe[name] = feature.apply(lambda x: int(x)**power)
            
    return poly_dataframe
```

```python
#Perform a ridge fit of a degree-16 polynomial using a "good" penalty strength

#"leave one out" (LOO) cross validation, which one can show approximates average mean square error (MSE). 
#As a result, choosing lambda to minimize the LOO error is equivalent to choosing lambda to minimize 
#an approximation to average MSE.

# LOO cross validation -- return the average MSE
def loo(data, deg, alpha_values):
    # Create polynomial features
    features = polynomial_features(data, deg)
    
    target = data["Y"]
    # Create as many folds for cross validatation as number of data points
    num_folds = len(data)
    
    folds = KFold(data.shape[0],num_folds)
    
    
    # for each value of l2_penalty, fit a model for each fold and compute average MSE
    alpha_mse = []
    min_mse = None
    best_alpha = None
    for alpha in alpha_values:
        
        next_mse = 0.0
        for train_set, validation_set in folds:
            
           
            # train model
            model = Ridge(alpha)
            
            train_predictors = (features.iloc[train_set,:])
            
    # The target we're using to train the algorithm.
            train_target = data["Y"].iloc[train_set]
            
    # Training the algorithm using the predictors and target.
            model.fit(train_predictors, train_target)
            
            
            # predict on validation set 
            y_test_predicted = model.predict(features.iloc[validation_set, :])
            
            # compute squared error
            next_mse += ((y_test_predicted-data["Y"].iloc[validation_set])**2).sum()
            
        # save squared error in list of MSE for each alpha
        next_mse = next_mse/num_folds
        alpha_mse.append(next_mse)
        if min_mse is None or next_mse < min_mse:
            min_mse = next_mse
            best_alpha = alpha
            
    return alpha_mse,best_alpha
print loo(data, 16, alpha_values)
```

```python
alpha_values = np.logspace(-4, 10, num=10)

alpha_mse,best_alpha = loo(data, 16, alpha_values)
```

```python
plt.plot(alpha_values,alpha_mse,'k-')
plt.xlabel('alpha')
plt.ylabel('LOO cross validation error')
plt.xscale('log')
plt.yscale('log')
```

### Lasso Regression

```python
def polynomial_lasso_regression(data, deg, l1_penalty):
    target=data['Y']
    model = linear_model.Lasso(alpha = l1_penalty)
    model.fit(polynomial_features(data,deg), target)
   
    return model
```

```python
for l1_penalty in [0.0001, 0.01, 0.1, 10]:
    model = polynomial_lasso_regression(data, deg=16, l1_penalty=l1_penalty)
    w = [model.intercept_] + list(model.coef_)

   
    
    
    non_zeros_number = len(np.nonzero(w)[0])
    
    
   
    
    
    print 'l1_penalty = %e' % l1_penalty
    print 'number of nonzeros = %r' % non_zeros_number 
    print_coefficients(model)
    print '\n'
    plt.figure()
    plot_poly_predictions(data,model)
    plt.title ('LASSO, ' + 'lambda '+ str(l1_penalty)  + ', number of nonzeros ' + str(non_zeros_number ))  
    
```
